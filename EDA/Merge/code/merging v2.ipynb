```python
import pandas as pd

df1 = pd.read_csv('wc_avg_apps.csv')
df2 = pd.read_csv('Final_dataframe.csv')

# country_map을 반대로 (영문 → 한글) 매핑
reverse_map = {v: k for k, v in country_map.items()}

# df1에 한글 이름 컬럼 추가
df1["팀"] = df1["Team"].map(reverse_map)

# merge (Year ↔ 년도, Team ↔ 영문국가명)
merged = pd.merge(
    df2,
    df1,
    left_on=["년도", "영문국가명"],
    right_on=["Year", "Team"],
    how="left"
)

merged
```

```python
# df1에는 reverse_map으로 한글 팀 안 붙이고 그냥 그대로 merge
merged = pd.merge(
    df2,
    df1,
    left_on=["년도", "팀"],
    right_on=["Year", "팀"],
    how="left"
)

# 불필요한 컬럼 정리
merged = merged.drop(columns=["Team", "Year"])  # 중복 키 제거
# 혹시 '팀_y'가 남았다면 제거
if "팀_y" in merged.columns:
    merged = merged.drop(columns=["팀_y"])
merged = merged.rename(columns={"팀_x": "팀"})

merged
```
