```python
import pandas as pd

df1 = pd.read_csv('avg_age.csv')
df2 = pd.read_csv('merged v2.csv')
reverse_map = {v: k for k, v in country_map.items()}

# df1에 한글 이름 컬럼 추가
df1["팀"] = df1["국가"].map(reverse_map)

merged = pd.merge(
    df2,
    df1,
    left_on=["년도", "팀"],
    right_on=["년도", "팀"],
    how="left"
)

# 불필요한 컬럼 정리
merged = merged.drop(columns=["국가"])  # 중복 키 제거
# 혹시 '팀_y'가 남았다면 제거
if "팀_y" in merged.columns:
    merged = merged.drop(columns=["팀_y"])
merged = merged.rename(columns={"팀_x": "팀"})

merged
```
