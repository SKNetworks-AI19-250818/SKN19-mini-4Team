```python
# 원본 df에 영어 국가명 추가
total_rank_df["country_eng"] = total_rank_df["팀"].map(country_map)

# win_df와 merge
total_rank_df = total_rank_df.merge(
    win_df[["wc_year", "country", "total_win_rate"]],
    left_on=["년도", "country_eng"],
    right_on=["wc_year", "country"],
    how="left",
).drop(columns=["wc_year", "country"])

# goal_df와 merge
total_rank_df = total_rank_df.merge(
    goal_df[["wc_year", "country", "total_gd_ratio"]],
    left_on=["년도", "country_eng"],
    right_on=["wc_year", "country"],
    how="left",
).drop(columns=["wc_year", "country"])

# 컬럼명 정리
total_rank_df = total_rank_df.rename(
    columns={"total_win_rate": "예선_승률", "total_gd_ratio": "예선_골득실률"}
)

print(total_rank_df)
```
