```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/results.csv')
df
```

```python
# 1990년 이후만
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'] >= "1990-01-01"]
```

```python
nations = pd.concat([df['home_team'],df['away_team']]).unique()
```

```python
len(nations)
```

```python
print(len(df['tournament'].unique()))
```

```python
# 세부 리그 대략적인 큰 범위로 축소
def categorize_tournament(t):
    if t == "FIFA World Cup":
        return "World Cup "
    elif t == "FIFA World Cup qualification":
        return "World Cup Qualifiers"
    elif "Euro" in t or "UEFA Nations League" in t:
        return "Europe"
    elif "AFC" in t or "Asian" in t or "EAFF" in t or "SAFF" in t or "AFF" in t or "CAFA" in t or "ASEAN" in t:
        return "Asia"
    elif "Africa" in t or "COSAFA" in t or "CECAFA" in t or "UNIFFAC" in t or "UDEAC" in t:
        return "Africa"
    elif "Copa" in t or "Gold Cup" in t or "CONCACAF" in t or "CONMEBOL" in t:
        return "Americas"
    elif "Oceania" in t or "Pacific" in t:
        return "Oceania"
    elif "Friendly" in t:
        return "Friendly"
    else:
        return "Other/Regional"

df["tournament_category"] = df["tournament"].apply(categorize_tournament)
```

```python
import pandas as pd
from tqdm import tqdm

wc_years = [2002, 2006, 2010, 2014, 2018, 2022]

# 지역 경기, 월드컵 예선, 친선전만 추가
valid_categories = [
    'Friendly', 'Africa', 'Europe', 'Asia', 'Americas', 'Oceania', 'World Cup Qualifiers'
]

results = []

for country in tqdm(nations, desc='Countries'):
    for wc_year in wc_years:
        start_year, end_year = wc_year - 3, wc_year + 2
        mask = (
            (df['date'].dt.year >= start_year) &
            (df['date'].dt.year <= end_year) &
            (df['tournament_category'].isin(valid_categories))
        )
        subset = df[mask]

        home_win = home_loss = home_draw = 0
        away_win = away_loss = away_draw = 0

        for _, row in subset.iterrows():
            if row['home_team'] == country or row['away_team'] == country:
                # 홈팀 neutral=False -> 홈 기록 아니면 어웨이로 기록
                if row['home_team'] == country and row['neutral'] == False:
                    if row['home_score'] > row['away_score']:
                        home_win += 1
                    elif row['home_score'] < row['away_score']:
                        home_loss += 1
                    else:
                        home_draw += 1
                else:
                    # 어웨이
                    if row['home_team'] == country:
                        if row['home_score'] < row['away_score']:
                            away_win += 1
                        elif row['home_score'] > row['away_score']:
                            away_loss += 1
                    elif row['away_team'] == country:
                        if row['away_score'] > row['home_score']:
                            away_win += 1
                        elif row['away_score'] < row['home_score']:
                            away_loss += 1
                        else:
                            away_draw += 1

        results.append({
            'wc_year': wc_year,
            'country': country,
            'home_win': home_win,
            'home_loss': home_loss,
            'home_draw': home_draw,
            'away_win': away_win,
            'away_loss': away_loss,
            'away_draw': away_draw
        })

stats_df = pd.DataFrame(results)
print(stats_df.head(10))
```

```python
cols = ['home_win', 'home_loss', 'home_draw', 'away_win', 'away_loss', 'away_draw']

# 모든 값이 0인 행 삭제
stats_df = stats_df[~(stats_df[cols] == 0).all(axis=1)]
```

```python
stats_df
```

```python
# 홈승률
stats_df['home_win_rate'] = stats_df['home_win'] / (
    stats_df['home_win'] + stats_df['home_loss'] + stats_df['home_draw']
)

# 어웨이승률
stats_df['away_win_rate'] = stats_df['away_win'] / (
    stats_df['away_win'] + stats_df['away_loss'] + stats_df['away_draw']
)

# 전체 승률
stats_df['total_win_rate'] = (stats_df['home_win'] + stats_df['away_win']) / (
    stats_df['home_win'] + stats_df['home_loss'] + stats_df['home_draw'] +
    stats_df['away_win'] + stats_df['away_loss'] + stats_df['away_draw']
)

stats_df[['home_win_rate', 'away_win_rate', 'total_win_rate']] = stats_df[['home_win_rate', 'away_win_rate', 'total_win_rate']].round(2)
```

```python
stats_df.to_csv('../created_data/team_winrate.csv',index=False,encoding = 'utf-8')
```

```python
import pandas as pd
from tqdm import tqdm

wc_years = [2002, 2006, 2010, 2014, 2018, 2022]

valid_categories = [
    'Friendly', 'Africa', 'Europe', 'Asia', 'Americas', 'Oceania', 'World Cup Qualifiers'
]

results = []

for country in tqdm(nations, desc='Countries'):
    for wc_year in wc_years:
        start_year, end_year = wc_year - 3, wc_year + 2
        mask = (
            (df['date'].dt.year >= start_year) &
            (df['date'].dt.year <= end_year) &
            (df['tournament_category'].isin(valid_categories))
        )
        subset = df[mask]

        # 득점/실점만 계산
        home_goals = home_conceded = 0
        away_goals = away_conceded = 0

        for _, row in subset.iterrows():
            if row['home_team'] == country or row['away_team'] == country:
                # neutral 이 Ture면 어웨이로 계산
                if row['home_team'] == country and row['neutral'] == False:
                    home_goals += row['home_score']
                    home_conceded += row['away_score']

                else:  # 어웨이
                    if row['home_team'] == country:
                        away_goals += row['home_score']
                        away_conceded += row['away_score']
                    elif row['away_team'] == country:
                        away_goals += row['away_score']
                        away_conceded += row['home_score']

        results.append({
            'wc_year': wc_year,
            'country': country,
            'home_goals': home_goals,
            'home_conceded': home_conceded,
            'away_goals': away_goals,
            'away_conceded': away_conceded,
        })

stats_df = pd.DataFrame(results)
stats_df
```

```python
# 모든 경우 골의 합산이 0인경우 제외
cols=['home_goals',	'home_conceded','away_goals','away_conceded']
stats_df = stats_df[~(stats_df[cols] == 0).all(axis=1)]
stats_df
```

```python
# 토탈 득점, 실점
stats_df['total_goals'] = stats_df['home_goals'] + stats_df['away_goals']
stats_df['total_conceded'] = stats_df['home_conceded'] + stats_df['away_conceded']

# 득실비
stats_df['home_gd_ratio'] = stats_df.apply(
    lambda x: x['home_goals'] / x['home_conceded'] if x['home_conceded'] > 0 else None, axis=1
)
stats_df['away_gd_ratio'] = stats_df.apply(
    lambda x: x['away_goals'] / x['away_conceded'] if x['away_conceded'] > 0 else None, axis=1
)
stats_df['total_gd_ratio'] = stats_df.apply(
    lambda x: x['total_goals'] / x['total_conceded'] if x['total_conceded'] > 0 else None, axis=1
)
```
